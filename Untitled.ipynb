```python
import matplotlib as plt
import numpy as np
import pandas as pd
import os
```

```python


#Loading Data

companies = pd.DataFrame.from_csv('companies.csv',0,';')
prices = pd.DataFrame.from_csv('prices.csv',0,';')
sectors = pd.DataFrame.from_csv('sectors.csv',0,';')

#companies
#prices
companies.reset_index(drop=False,inplace=True)
prices.reset_index(drop=False,inplace=True)
print('Dimensions of the Company Dataset'+ str(companies.shape))
print('')
print(list(companies))
print('-------------------------------------------------------------------------')
print('Dimensions of the Prices Dataset'+ str(prices.shape))
print('')
print(list(prices))
print('-------------------------------------------------------------------------')
print('Dimensions of the Sector Dataset'+ str(sectors.shape))
print('')
print(list(sectors))

companies
```

```python
#Combining the Data into dictionaries

mess = pd.merge(companies, prices, on='ISIN_CODE', how='outer')
mess.fillna(value='unknown',inplace=True)
mess.to_csv('companies_prices.csv')
df = mess.set_index('NAME')

```

```python
df_split = {x: df.loc[x]  for x in df.index.unique()}
```

```python
path = 'company_dataframes/'
for i in df.index.unique():
    filename = str(i)+'.csv'
    df_split[i].to_csv(path + filename)

```
